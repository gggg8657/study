```python
#1. Greedy Policy
def clip(x):
    return min(3, max(0, x))

def Greedy_Policy(epoch):
    G = [[-0.0]*4 for _ in range(4)]
    def update():
        G_memory = [[-0.0]*4 for _ in range(4)]
        for i in range(4):
            for j in range(4):
                if i==j and (i==0 or i==3):
                    continue

                summation = []

                for i2, j2 in [[i+1,j],[i,j+1],[i-1,j],[i,j-1]]:
                    i2 = clip(i2)
                    j2 = clip(j2)
                    summation.append(G[i2][j2]/4)

                G_memory[i][j] = -1 + sum(summation)

        return G_memory


    for i in range(epoch):
        G = update()

    print(f"\nGreedy_Policy {epoch} times iteration")
    for line in G:
        print(list(map(lambda x: round(x, 1), line)))

def Value_Iteration(epoch):
    G = [[-0.0]*4 for _ in range(4)]
    def update():
        G_memory = [[-0.0]*4 for _ in range(4)]
        for i in range(4):
            for j in range(4):
                if i==j==0 or i==j==3:
                    continue

                summation = []

                for i2, j2 in [[i+1,j],[i,j+1],[i-1,j],[i,j-1]]:
                    i2, j2 = clip(i2), clip(j2)
                    j2 = clip(j2)
                    summation.append(G[i2][j2])

                G_memory[i][j] = -1 + max(summation)

        return G_memory


    for i in range(epoch):
        G = update()

    print(f"\nValue_Iteration {epoch} times iteration")
    for line in G:
        print(list(map(lambda x: round(x, 1), line)))

if __name__ == '__main__':
    epoch = 100
    Greedy_Policy(epoch=epoch)
    Value_Iteration(epoch=epoch)
```
